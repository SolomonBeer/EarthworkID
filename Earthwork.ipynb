```python
def find_edges(image_name):
    image = Image.open(image_name)
    image = image.convert(mode="P")
    image_array = np.array(image, dtype=np.uint8)
    edges = filters.sobel(image_array)
    return(edges)
```

```python
badbury_rings = find_edges("Badbury Rings Blandford BW.png")
cadbury_castle = find_edges("Cadbury Castle Sherbourne BW.png")
maiden_castle = find_edges("Maiden Castle Dorchester BW.png")
hambledon_hill = find_edges("Hambledon Hill Blandford BW.png")
hod_hill = find_edges("Hod Hill Blandford BW.png")
cadsonbury_hill = find_edges("Cadson Bury Hill Plymouth BW.png")
cranbrook_castle = find_edges("Cranbrook Castle Exeter BW.png")
eggardon_hill = find_edges("Eggardon Bridport BW.png")
pilsdon_pen = find_edges("Pilsdon Pen BW.png")
prestonbury_castle = find_edges("Prestonbury Castle Exeter BW.png")
barbury_castle = find_edges("Barbury Castle BW.png")
battlesbury_hillfort = find_edges("Battlesbury Hillfort BW.png")
beacon_hill = find_edges("Beacon Hill BW.png")
british_camp = find_edges("British Camp Hill Fort BW.png")
cley_hill = find_edges("Cley Hill BW.png")
ladle_hill = find_edges("Ladle Hill BW.png")
liddington_castle = find_edges("Liddington Castle BW.png")
middle_hill = find_edges("Middle Hill BW.png")
old_winchester = find_edges("Old Winchester Hill BW.png")
oldbury_castle = find_edges("Oldbury Castle BW.png")
scratchbury_hillfort = find_edges("Scratchbury Hillfort BW.png")
uffington_castle = find_edges("Uffington Castle BW.png")
```
