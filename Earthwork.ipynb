```python
def find_edges(image_name):
    image = Image.open(image_name)
    image = image.convert(mode="P")
    image_array = np.array(image, dtype=np.uint8)
    edges = filters.sobel(image_array)
    return(edges)
```

```python
badbury_rings = find_edges("Badbury Rings Blandford BW.png")
cadbury_castle = find_edges("Cadbury Castle Sherbourne BW.png")
maiden_castle = find_edges("Maiden Castle Dorchester BW.png")
hambledon_hill = find_edges("Hambledon Hill Blandford BW.png")
```

```python
io.imshow(badbury_rings)
```

```python
model = NMF(n_components=8, init='random', random_state=0, max_iter=2000)
```

```python
#need to NMF on the n different sites rather than on just one? though some promise in the 8 columns seen
W = model.fit_transform(badbury_rings)
H = model.components_
```
